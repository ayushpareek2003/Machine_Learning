## TYPES OF GRAPH

1) SCATTER

```python
#IT IS A TYPE OF GRAPH IN WHICH WE PLOT BY POINTS WITHOUT JOING THEM
```

```python
plt.scatter(x,y**2);
```

```python
#we can use scatter everywhere 
```

2) Bar

```python
#second form is bar thorugh this we can present data in bar graph
```

```python
plt.bar(x,x**2);
```

```python
#another form is horizontal bar
plt.barh(x,y);
```

3) Hist

```python
plt.hist([2,10,45],[5,26,47]);
```

```python
#these were basic graph type 
#we can draw more than 1 type of graph in same graph
st3,ar=plt.subplots()
```

```python
ar.plot(x,y);
ar.scatter(x,y);
ar.barh(x,y);
st3
```

```python
#now we see making graph from csv data
#im importing a file which i will attacth in folder
car=pd.read_csv("52  car-sales.csv")
car
```

```python
#it will automaticaly plot integer data but this is pandas inbuilt function not from matplpotlib
car.plot()
```

```python
#using matplotlib
plt.plot(car["Doors"]);
```

```python
ya=np.random.randint(0,10,10)
plt.scatter(car["Price"],ya);
```

```python
#cumsum is used to get cumulative sum

car["Total"]=car["Price"].astype(int).cumsum()
car
#we will get a error now 
```

we are getting a error beacuse our price coloumn is not in right form it must be in int with base 10

```python
#for this change its type
```

```python
car["Price"]=car["Price"].str.replace('[\$\,\.]','')
```

```python
#you can dollar sign and , and. is removed
car
```

```python
car["Total"]=car["Price"].astype(int).cumsum()
```

```python
car
```

```python
plt.scatter(car["Price"],car["Total"]);
```

```python
#another way of doing it is
car.plot("Price","Total",kind="bar");
```

```python
#bins in hist function is size of the bar we can decrease the width by increasing its value
car.hist(bins=50);
```
