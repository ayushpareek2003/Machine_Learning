Now we import csv file and perform some function on them(function on data frame and csv file are same )

```python
#importing a csv file(i also included a csv file in repository)
r=pd.read_csv("52  car-sales.csv")
r
```

```python
#operations
#accesing a row ,for this we will use two function iloc(location based) and loc(index based)
r.iloc[2]
```

```python
r.loc[2]
```

```python
#info about dataframe 
r.dtypes
```

```python
#basic results about dataframe
r.describe
```

```python
#now we try to replace values in data frame 
r
```

```python
#we will change doors from 4 to 3
r["Doors"]=r["Doors"].replace(4,3)
r
```

```python
#another way
r["Doors"].replace(3,4,inplace=True)
r
```

```python
#get data with some condition
r[r["Doors"]>2]
```

```python
r[r["Colour"]=="White"]
```

```python
#getting some upper rows of data frame
r.head()
```

```python
#getting some lower rows of data frame
r.tail()
```

```python
#Adding a coloumn to get doors as cumlative sum
r["Total"]=r["Doors"].astype(int).cumsum()
r
```

```python
#plotting graphs
r.plot()
```

```python
r["Doors"].plot()
```

```python
r
```

```python
#now we will see a example for how to remove specific charcter from columns (Ex=removing decimal from price column)
r["Price"]=r["Price"].str.replace('[\.]','')
```
